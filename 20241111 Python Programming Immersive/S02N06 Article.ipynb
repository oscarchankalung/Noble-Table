```python
file = open(r"article.txt", "r", encoding="utf8")
data = file.read().lower()
data
```

```python
# cleaning
new_data = ""

for char in data:
    if char.isalpha() or char.isspace():
        new_data += char

data = new_data.split()
data
```

```python
word_count = {}
get_number = lambda t : t[1] # helper function for sort

for word in data:
    word_count[word] = word_count.get(word, 0) + 1

word_count = list(word_count.items())
word_count.sort(key = get_number, reverse = True)
word_count[:10]
```
