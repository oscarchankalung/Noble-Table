```python
url = "https://rickandmortyapi.com/api/character"

response = requests.get(url)
data = response.json()
data
```

```python
# print all names and status of all chars
```

```python
data.keys()
```

```python
data["results"][0].keys()
```

```python
name_and_species = [(result["name"], result["species"]) for result in data["results"]]
name_and_species
```

```python
species_count = {}

for name, species in name_and_species:
    species_count[species] = species_count.get(species, 0) + 1

species_count
```

```python
labels = list(species_count.keys())
sizes = list(species_count.values())
explode = [0.2,0]
autopct = "%.2f%%"
shadow = True
colors = ["#DFFF00", "#6495ED"]

plt.pie(sizes, labels=labels, explode=explode, autopct=autopct, shadow=shadow, colors=colors)
plt.title("Percentage of Species")
plt.savefig("percentage_of_species_pie_chart.png")
```
