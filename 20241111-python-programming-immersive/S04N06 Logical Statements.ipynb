```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(list(range (1,11)), columns= ["Numbers"])
df
```

```python
df["Lambda"] = df["Numbers"].apply(lambda x : "Even" if x%2 == 0 else "Odd" )
df
```

```python
df["Where"] = np.where(df["Numbers"]%2==0, "Even", "Odd")
df
```

```python
df.loc[df["Numbers"]%2==0 , "Location"] = "Even"
df.loc[df["Numbers"]%2!=0 , "Location"] = "Odd"
df
```

```python
df["New"] = df.apply(lambda d : d["Numbers"] * 100 if d["Location"] == "Even" else d["Numbers"] * 10 , axis="columns")
df
```

```python
store1 = pd.DataFrame()
store1["Product1"] = ["Mac", "IPad", "IPhone", "IWatch", "TV"]
store1["Price1"] = [2000, 800, 700, 400, 300]
store1
```

```python
store2 = pd.DataFrame()
store2["Product2"] = ["Mac" ,"IPad", "IPhone", "IWatch", "TV" ]
store2["Price2"] = [1900, 750, 700, 350, 300]
store2
```

```python
# Compare prices "Price1" vs "Price2" if same True if not False use np.where
```

```python
# solution

store1["Same Price"] = np.where(store1["Price1"] == store2["Price2"], True, False)
store1
```

```python
# if price is not the same, decrease Price1 by 10% and else leave price as it is
# save results under col name Sale, use np.where
```

```python
store1["Sale"] = np.where(store1["Same Price"], store1["Price1"],  store1["Price1"] * 0.9)
store1
```

```python
# solve this problem with loc
# if price is 700 decrease by 20%
```

```python
store1.loc[store1["Price1"] > store2["Price2"], ["Mega Sale"]] = store1["Price1"] * 0.9
store1.loc[store1["Price1"] <= store2["Price2"], ["Mega Sale"]] = store1["Price1"]
store1.loc[store1["Price1"] == 700, ["Mega Sale"]] = store1["Price1"] * 0.8
store1
```

```python
# use apply lambda if Same Price leave the Pricel as it is otherwise decrease by 10%
```

```python
store1["Xmas Sale"] = store1.apply(lambda d : d["Price1"] if d["Same Price"] else d["Price1"] * 0.9, axis=1)
store1
```
